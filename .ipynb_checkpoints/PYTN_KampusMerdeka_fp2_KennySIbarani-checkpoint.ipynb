# Data Preprocessing

```python
# Dataset
weather.head()
```

```python
# Memisahkan kolom type kategori
weather_categ = [var for var in weather.columns if weather[var].dtype=='object']
weather_categ = weather[weather_categ]

weather_categ.head()
```

```python
# Label Encoder data kategorikal
Lenc = LabelEncoder()

weather_categ["location"] = Lenc.fit_transform(weather_categ["location"])
weather_categ["windgustdir"] = Lenc.fit_transform(weather_categ["windgustdir"])
weather_categ["winddir9am"] = Lenc.fit_transform(weather_categ["winddir9am"])
weather_categ["winddir3pm"] = Lenc.fit_transform(weather_categ["winddir3pm"])

weather_categ.head()
```

```python
# Mengganti kolom kategori dengan label encoder
weather["location"] = weather_categ["location"]
weather["windgustdir"] = weather_categ["windgustdir"]
weather["winddir9am"] = weather_categ["winddir9am"]
weather["winddir3pm"] = weather_categ["winddir3pm"]
```

```python
# Menampilkan dataset yang sudah menjadi data numerik
weather_new = weather
weather_new
```

    Pada kolom yang bertipe data objek harus di ganti dengan data numerik agar dapat dianalisis. 

```python
# Menghapus kolom yang tidak terlalu berkorelasi
drop_col = ['mintemp', 'maxtemp', 'sunshine', 'cloud9am', 'cloud3pm', 'winddir9am', 'winddir3pm']
weather_new = weather_new.drop(columns = drop_col)
weather_new.head()
```

Beberapa kolom pada dataset yang kurang berkorelasi dapat dihapus karena dapat diwakilkan dengan kolom lain yang relevan, guna memprediksi hujan yang akan turun esok hari. Membuat variabel untuk data train dan data test. Untuk variabel x adalah data pendukung untuk memprediksi dan data y adalah data yang akan di prediksi yaitu 'raintomorrow'.

```python
# Membuat variabel x untuk semua kolom kecuali kolom 'raintomorrow'
x = weather_new.drop(columns=['raintomorrow'])

# Membuat variabel y untuk kolom 'raintomorrow'
y = weather_new['raintomorrow']
```

```python
# Memisahkan data menjadi train set dan test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Pendefinisian Model

LogisticRegression() adalah sebuah kelas dalam library atau modul yang digunakan untuk mengimplementasikan algoritma logistic regression. Logistic regression adalah metode yang digunakan untuk klasifikasi dan estimasi probabilitas berdasarkan hubungan linier antara variabel independen dan variabel dependen biner atau kategori. Objek LogisticRegression() dapat digunakan untuk melatih model logistic regression pada data dan melakukan prediksi berdasarkan model yang telah dilatih.

SVC() adalah sebuah kelas dalam library atau modul yang digunakan untuk mengimplementasikan algoritma Support Vector Machine (SVM) untuk klasifikasi dan regresi. SVM merupakan metode pembelajaran mesin yang mencari hyperplane optimal untuk memisahkan dua kelas data dengan margin maksimal. Objek SVC() dapat digunakan untuk melatih model SVM pada data dan melakukan prediksi berdasarkan model yang telah dilatih.

```python
logreg = LogisticRegression()
svm = SVC()
```

# Pelatihan Model

```python
logreg.fit(x_train, y_train)
svm.fit(x_train, y_train)
```

```python
# Menampilkan jumlah data dalam setiap set
print("Jumlah data dalam train set:", len(x_train))
print("Jumlah data dalam test set:", len(x_test))
```

# Evaluasi Model

```python
log_model = LogisticRegression().fit(x_train,y_train)
y_pred_train = log_model.predict(x_train)
y_pred_test = log_model.predict(x_test)
```

```python
confusion_matrix_data = pd.DataFrame((confusion_matrix(y_train, y_pred_train)), ('No Rain', 'Rain'), ('No Rain', 'Rain'))

plt.figure()
heatmap = sns.heatmap(confusion_matrix_data, annot = True, annot_kws={'size': 14}, fmt='d', cmap='YlGnBu')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)

plt.title('Confusion Matrix for Training Model\n(Logistic Regression)', fontsize=18, color='darkblue')
plt.ylabel('True label', fontsize=14)
plt.xlabel('Predict label', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
print('Classification Report Training Model (Logistic Regression) : ')
print(classification_report(y_train, y_pred_train))
```

```python
confusion_matrix_data = pd.DataFrame((confusion_matrix(y_test, y_pred_test)), ('No Rain', 'Rain'), ('No Rain', 'Rain'))

plt.figure()
heatmap = sns.heatmap(confusion_matrix_data, annot = True, annot_kws={'size': 14}, fmt='d', cmap='YlGnBu')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)

plt.title('Confusion Matrix for Test Model\n(Logistic Regression)', fontsize=18, color='darkblue')
plt.ylabel('True label', fontsize=14)
plt.xlabel('Predict label', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
print('Classification Report Test Model (Logistic Regression) : ')
print(classification_report(y_test, y_pred_test))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

print("Test Score:\n",log_model.score(x_train,y_train)*100)
print("Mean Squared Error:\n",mean_squared_error(y_test,y_pred_test))
print("R2 score is:\n",r2_score(y_test,y_pred_test))
```

Test Score: 84.87%
Training Score menunjukkan tingkat akurasi model pada data pelatihan. Angka 84.87% menunjukkan bahwa model memiliki tingkat keakuratan sebesar itu dalam memprediksi data pelatihan.

Mean Squared Error: 0.1526
Mean Squared Error (MSE) adalah metrik yang mengukur sejauh mana kesalahan prediksi model dengan nilai sebenarnya dalam kuadrat. Semakin rendah MSE, semakin baik kualitas prediksi model. Dalam kasus ini, nilai MSE sebesar 0.1526 menunjukkan tingkat kesalahan prediksi yang relatif rendah.

R2 Score: 0.1185
R2 Score, atau koefisien determinasi, adalah metrik yang mengukur seberapa baik model dapat menjelaskan variasi dalam data. Nilai R2 Score berkisar antara 0 hingga 1, dan semakin tinggi nilainya, semakin baik model dapat menjelaskan variasi data. Dalam kasus ini, nilai R2 Score sebesar 0.1185 menunjukkan bahwa model memiliki tingkat penjelasan yang rendah terhadap variasi dalam data.

```python
log_accuracy = (accuracy_score(y_test,y_pred_test)*100)
print(log_accuracy)
```
