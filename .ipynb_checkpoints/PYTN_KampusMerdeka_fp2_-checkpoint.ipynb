    Keterangan:
    1. Beberapa kolom memiliki outlier yang jelas (mis., nilai Rainfall maks adalah 371,0 meskipun persentil ke-75 adalah   0,8)
    2. Tidak melihat nilai apa pun yang langsung menimbulkan kekhawatiran (seperti nilai negatif untuk Curah Hujan minimum)
    
    Untuk mendapatkan pemahaman yang lebih baik tentang data dan menangkap nilai placeholder apa pun yang mungkin tidak     muncul dalam statistik ringkasan, kami juga ingin memeriksa lima nilai paling teratas untuk setiap kolom.

```python
for col in weather.columns:
    print('\n')
    print(col)
    print('-'*15)
    print(weather[col].value_counts(normalize=True).head())
```

    Sekitar 64% pengamatan memiliki nilai 0 untuk Curah Hujan, sekitar 77,5% hari tidak memiliki curah hujan menurut dua    kolom terakhir. Perbedaan ini kemungkinan karena perbedaan jumlah nilai yang hilang untuk setiap kolom.
    Kolom RainToday dan RainTomorrow harus diubah menjadi 0 dan 1 untuk manipulasi yang lebih mudah
    Jelajahi lebih lanjut kolom Tanggal dengan mengonversi kolom bernama "Date" dalam pandas DataFrame (`df`) dari format   string ke format datetime.

```python
weather["date"] = pd.to_datetime(weather["date"], format = "%Y-%m-%d", errors = "coerce")
```

```python
weather['date'].info()
```

```python
weather.date.value_counts()
```

```python
weather.location.nunique()
```

    Jumlah maksimum pengamatan untuk tanggal tertentu sejajar dengan jumlah lokasi unik dalam kumpulan data. Ini secara     intuitif masuk akal karena setiap stasiun cuaca di lokasi yang berbeda akan melaporkan data mereka sendiri untuk hari    tertentu. Menyesuaikan kolom RainToday dan RainTomorrow:

```python
weather.raintoday = weather.raintoday.map({'No': 0, 'Yes': 1})
weather.raintoday.value_counts(normalize=True)
```

```python
weather.raintomorrow = weather.raintomorrow.map({'No': 0, 'Yes': 1})
weather.raintomorrow.value_counts(normalize=True)
```
